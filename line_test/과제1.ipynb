```python
import os
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load
with open('data.pickle', 'rb') as f:
    data = pickle.load(f)
```

```python
image_list = []
for i in range(10):
    image_list.append(data[str(i)])
```

```python
def cal_img_mean_std(image):
    image_mean_std = []
    for c in range(3):
        m = 0
        for y in range(len(image)):
            for x in range(len(image[0])):
                m += image[y][x][c]
        mean = m/(len(image)*len(image[0]))

        s = 0
        for y in range(len(image)):
            for x in range(len(image[0])):
                s += (image[y][x][c]-mean)**2
        std = (s/(len(image)*len(image[0])))**0.5

        image_mean_std.append((mean, std))
    return image_mean_std

def cal_mean_std(image_list):
    means = [0, 0, 0]
    stds = [0, 0, 0]
    for image in image_list:
        image_mean_std = cal_img_mean_std(image)
        for c in range(3):
            mean, std = image_mean_std[c]
            means[c] += mean
            stds[c] += std
    
    for c in range(3):
        means[c] /= len(image_list)*255
        stds[c] /= len(image_list)*255
    return means, stds

def image_normalization(image, means, stds):
    for c in range(3):
        for y in range(len(image)):
            for x in range(len(image[0])):
                image[y][x][c] = (image[y][x][c] - means[c])/stds[c]
    return image
```

```python
means, stds = cal_mean_std(image_list)
print(means)
print(stds)
for image in image_list:
    plt.imshow(np.array(image))
    plt.show()
    
    image = image_normalization(image, means, stds)

    plt.imshow(np.array(image))
    plt.show()
```
